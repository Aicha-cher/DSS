```python
# example data
import numpy as np
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)

# example error bar values that vary with x-position
error = 0.1 + 0.2 * x

fig, (ax0, ax1) = plt.subplots(nrows=2, sharex=True)
ax0.errorbar(x, y, yerr=error, fmt='-o')
ax0.set_title('variable, symmetric error')

# error bar values w/ different -/+ errors that
# also vary with the x-position
lower_error = 0.4 * error
upper_error = error
asymmetric_error = [lower_error, upper_error]

ax1.errorbar(x, y, xerr=asymmetric_error, fmt='o')
ax1.set_title('variable, asymmetric error')
ax1.set_yscale('log')
plt.show()
```

## Task b)
#### Plot the boxplots of median earnings for the top 5 categories (as measured by median earnings). Then add the raw sample points to the plot using a suitable plot type. Make it nice looking and prepare the plot for publication. Then save it in a suitable file format to load it into Inkscape.

```python
top_5_df = df.sort_values(by=['Median'], ascending=False)[:5]
```

```python
with paperStyle(font_size=8):
    sns.boxplot(top_5_df,
                x='Major_category',
                y='Median',
                color='#ACAFFA',

                width=0.5,
                fliersize=2)

    # Plot data points
    sns.stripplot(top_5_df,
                x='Major_category',
                y='Median',
                size=5,
                jitter=False,
                legend=False,
                color='#9699e4',
                edgecolor='gray',
                linewidth=1.75)
    sns.despine()
    plt.ylabel('Median Salary')
    plt.xlabel('University Major Categories')
    plt.savefig('./task2.svg', bbox_inches='tight')
```

```python
unemp_df = df.sort_values('Unemployment_rate', ascending=False).head(10)
unemp_df.Unemployment_rate = unemp_df.Unemployment_rate * 100
unemp_df.Major_category = unemp_df.Major_category.str.casefold()
with paperStyle(font_size=8):
    sns.violinplot(unemp_df, x='Major_category', y='Unemployment_rate',
                    palette='pastel',
                    width=0.25)
    plt.xticks(rotation=45, ha='right')
    sns.swarmplot(unemp_df, x='Major_category', y='Unemployment_rate',
                size=5,
                palette='pastel',
                edgecolor='gray',
                linewidth=2,
                legend=False)
    plt.ylabel('Unemployment Rate %')
    plt.xlabel('University Major Categories')
    plt.ylim([0, 25])
    sns.despine()
    plt.savefig('./task3.svg', bbox_inches='tight')
```
